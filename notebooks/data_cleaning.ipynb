```python
orders = orders[orders["order_status"] == "delivered"].copy()
```

```python
orders.isnull().sum()
```

```python
orders = orders.dropna(subset=["order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date"])
```

```python
orders.isnull().sum()
```

```python
products = products.dropna(subset=["product_category_name"])
```

```python
products.isnull().sum()
```

```python
products = products.dropna()
```

```python
print(orders.isnull().sum())
print(products.isnull().sum())
```

```python
orders.to_csv("../data/orders_clean.csv", index = False)

products.to_csv("../data/products_clean.csv", index = False)
```

```python
df = (
    orders
    .merge(customers, on = "customer_id", how = "inner")
    .merge(items, on = "order_id", how = "inner")
    .merge(payments, on = "order_id", how = "inner")
    .merge(products, on = "product_id", how = "inner")
)
```
